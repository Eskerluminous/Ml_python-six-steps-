```python
num_value=[value for value in df.columns if df[value].dtypes not in ['0','object'] ]
print("Total number of numerical index:",len(num_value))
df[num_value].head()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("Car price")
sns.distplot(df["price"])
plt.subplot(1,2,2)
plt.title("car price")
sns.boxplot(y=df["price"])
```

```python
for value in num_value:
  if value not in ["car_ID"]:
    plt.scatter(y=df["price"],x=df[value])
    plt.ylabel(["price"])
    plt.xlabel(value)
    plt.show()
```
